# 將程式碼包成 function

```
import yfinance as yf

def crawl_price(stock_id):
  df = yf.Ticker(stock_id).history(period="max")
  return df
```

# 調用 function

```
df = crawl_price("2354.TW")
df.Close.plot()
```
